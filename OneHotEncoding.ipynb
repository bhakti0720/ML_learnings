## OneHotEncoding using Pandas
So in ur dataset, the brand column has 32 unique values.

That's a huge value 😶
  
which can increase dimensionality if we directly apply encoding.

This can lead to:
- High memory usage  
- Increased model complexity  
- Risk of overfitting  

👽Solution: Threshold-based Grouping

To reduce dimensionality, we apply a **threshold**.

We keep only those brands that appear **more than 100 times**  
All other less frequent brands are grouped into a single category: **"Uncommon"**




```python
counts = df['brand'].value_counts()
print(counts)
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
print(repl)
```

This repl variable stores all the uncommon cars brand

It will stores all this brands in array

Now we are using the Pandas get_dummies class to encode the brand column

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'),dtype=int).sample(5)
```

Doing the one-hot encoding for the

## K-1 One-Hot Encoding for 'fuel' and 'owner'

### What is K-1 One-Hot Encoding?

In One-Hot Encoding, if a column has **K categories**, we create **K-1 dummy variables** instead of K.


### Why do we use K-1 encoding?

To avoid **Dummy Variable Trap (Multicollinearity)**.

When we create K dummy variables:
- One column can be predicted using others  
- This creates **redundant information**  <br>
(Note the above point)

Example:
If we create 4 dummy variables:
- Diesel  
- Petrol  
- CNG  
- LPG  

One column becomes redundant because:
If we know 3 values, we can automatically determine the 4th.
so we drop 1 column from it (i.e the k-1)


###  What is the Dummy Variable Trap? 🪤

It is a situation where:
- One feature is **perfectly dependent** on others (idealy i/p columns should be independent of each other )
- This causes problems in models like **Linear Regression**  


###  How K-1 solves this?

 By dropping one category:
- We remove redundancy  
- Model becomes stable and efficient  



```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True, dtype=int)
```

## OneHotEncoding using Sklearn

Generally we ues encoding by using Sklearn in Ml

you might think why ???<br>
bcz
- sklearn encoders (like `OneHotEncoder`) learn categories during training (`fit`)
- Apply the same transformation on test data (`transform`)

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',dtype=np.int32)
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_test_new = ohe.transform(X_test[['fuel','owner']])
X_train_new.shape
```

Here we are getting the final 7 columns as a result <br>
bcz:
- 3 columns from the fules (after removing 1 from the total of 4)
- 4 columns from the owner (after removing 1 from the total of 5)
